```python
#Import nltk dictonary to remove repeative words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
#Now to match the numbers of words from movies or rows i chose vectorization method
# Vectorization will match or pick the nearest first 5 vectors as per the given input by the user.
# Make sure always remove stop words like (is,are,to) words which have no use
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
#After Stem function to Analysis the similarity we need to calculate the distance between vectors
#Make sure we have to calculate Cosine distance in angles not the Euclidean distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
#Similarity would always run in diagonally.
```

```python
#to fix the similarities between movies , we need to use the function called Enumerate
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:

        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
